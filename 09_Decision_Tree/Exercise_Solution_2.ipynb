```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:\\Programming\\Python\\py-master\\ML\\9_decision_tree\\Exercise\\titanic.csv')
df
```

```python
fill=round(df['Age'].mean())
df['Age']=round(df['Age'].interpolate(method='linear',axis=0))
df.head(20)
```

```python
le=LabelEncoder()
df['label_sex']=pd.DataFrame(le.fit_transform(df['Sex']))
x=df[['Pclass','label_sex','Age','Fare']]
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y,train_size=0.9)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
